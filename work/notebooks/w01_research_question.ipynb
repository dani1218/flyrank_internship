## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

```python
# Quick shape check on the dataset to confirm it is usable
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dani1218/flyrank_internship/main/data/fake_job_postings.csv')
print('Dataset rows:', len(df))
print('Columns:', df.columns.tolist())
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# Check how many postings include a salary range vs not — one of the key features
has_salary = df['salary_range'].notna().sum()
no_salary = df['salary_range'].isna().sum()
has_logo = df['has_company_logo'].sum()
no_logo = len(df) - has_logo

print(f'Postings with salary range:    {has_salary} ({has_salary/len(df)*100:.1f}%)')
print(f'Postings without salary range: {no_salary} ({no_salary/len(df)*100:.1f}%)')
print(f'Postings with company logo:    {has_logo} ({has_logo/len(df)*100:.1f}%)')
print(f'Postings without company logo: {no_logo}  ({no_logo/len(df)*100:.1f}%)')
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# Description length by employment type
df['desc_word_count'] = df['description'].fillna('').apply(lambda x: len(x.split()))

ft_avg = df[df['employment_type'] == 'Full-time']['desc_word_count'].mean()
pt_avg = df[df['employment_type'] == 'Part-time']['desc_word_count'].mean()
missing_desc = df['description'].isna().mean() * 100

print(f'Average description length (full-time):  {ft_avg:.0f} words')
print(f'Average description length (part-time):  {pt_avg:.0f} words')
print(f'Missing description rate: {missing_desc:.1f}%')
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

```python
# Confirm scope: industry spread and language skew
unique_industries = df['industry'].nunique()
top_industries = ' | '.join(df['industry'].value_counts().head(3).index.tolist())

# Rough non-English estimate: postings where description contains non-ASCII characters
non_english = df['description'].fillna('').apply(
    lambda x: any(ord(c) > 127 for c in x)
).mean() * 100

print('Scope check passed.')
print(f'Unique industries in dataset: {unique_industries}')
print(f'Top 3 industries: {top_industries}')
print(f'Non-English postings estimated: ~{non_english:.1f}% — results are bounded to English-language context.')
```
